## - Exercici 4
Implementa manualment una funció que calculi el coeficient de correlació. Informa-te'n sobre els seus usos i interpretació.

```python
import numpy as np

x= np.array ([1.89,1.67,1.25,1.04])
y= np.array([98,76,25,16])
print ('x: ',x,'y: ',y)

media_x= np.sum(x) / len(x) #media x
print ('media x: ',media_x)
media_y= np.sum(y) / len(y) #media y
print ('media y: ',media_y)

def dif_x (x,media_x): 
    dif_x = x - media_x
    return dif_x

def dif_y (y,media_y): 
    dif_y = y - media_y
    return dif_y

def prod_dif (dif_y , dif_x):
    prod_dif = dif_x * dif_y
    return prod_dif

difx = dif_x(x,media_x)
print('diferencia_x_mediax: ', difx)

dify = dif_y(y,media_y)
print('diferencia_y_mediay: ', difx)

prod = prod_dif(dify , difx)
print ("prod_dif: ")
print (prod_dif(dify , difx))

total =np.sum (prod)
numerador = total / len (prod)
print (numerador)

vx = np.var(x) #varianza x
print ('varianza x: ',vx)
vy = np.var(y) #varianza y
print ('varianza y: ',vy)
vx_vy = vx* vy #Producto de varianza x * varianza y
print (vx_vy)

denominador= (vx_vy) ** (0.5) #Raiz cuadrada de vx * vy
print (denominador)
r =numerador / denominador
print ('coeficiente de correlación manual: ',r)

# Calculamos el nivel de correlación
matriz_corr = np.corrcoef(x, y)
print ('Nivel de correlación: ', matriz_corr)

np.corrcoef(x, y)[0,1]

```

En este caso, el valor mayor que 0 indica que existe una correlación positiva. Las variables estarían asociadas en sentido directo. Cuanto más cerca de +1, más alta es su asociación. Es una relación lineal casi positiva perfecta.

Una forma de cuantificar la relación entre dos variables es utilizar el coeficiente de correlación de Pearson , que es una medida de la asociación lineal entre dos variables.
Podemos definir el coeficiente de correlación de Pearson comoun índice que puede utilizarse para medir el grado de relación de dos variables siempre y cuando ambas sean cuantitativas y continuas.
